```python
import pandas as pd
data = pd.read_csv("./data/Spotify_Youtube.csv")
```

```python
data.columns
data.head()
```

```python
data.info()
data.isnull().sum()/len(data)
```

```python
duplicates = data.duplicated(subset=["Artist", "Track"], keep=False)
duplicate_rows = data[duplicates]
print(duplicate_rows)
```

```python
data.drop_duplicates(subset=["Artist", "Track"], inplace=True)
```

```python
data_clean = data.dropna(subset=['Title', 'Danceability', 'Energy', 'Key', 'Loudness', 'Speechiness',
       'Acousticness', 'Instrumentalness', 'Liveness', 'Valence', 'Tempo'], axis=0)

data_clean.head()
data_clean.isnull().sum()
```

```python
data_clean.to_csv('./data/data_clean.csv',index=False)
```
